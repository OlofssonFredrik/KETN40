```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
F = 1
V1 = 5
V2 = 10
cin = 1
A = np.array([
            [-F/V1,0],
            [F/V2,-F/V2]
            ])
B = np.array([F/V1,0])
def model(t, c):
    dcdt = A@c + B*cin
    return dcdt
```

```python
c0 = np.array([0,0])
t_span = [0,60.]
sol = solve_ivp(model, t_span, c0, atol=1e-9, rtol=1e-6)
t = sol.t
c = sol.y.T
```

```python
plt.figure()
plt.plot(t, c[:,0], label=r'$c_1$')
plt.plot(t, c[:,1], label=r'$c_2$')
plt.legend()
```
