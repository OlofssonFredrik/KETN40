```python
from scipy.optimize import differential_evolution
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rosenbrock(x, print_vals = False):
    x1, x2 = x 
    Q = 100*(x2 - x1**2)**2 + (1 - x1)**2
    if print_vals == (2,):
        print(f'Current x1: {x1:.5} Current x2: {x1:.5} Current Q: {Q:.5}')
    return Q

def six_humps(x, print_vals = False):
    x1, x2 = x
    Q = -4*( (4 - 2.1*x1**2 + x1**4/3)*x1**2 + x1*x2 + (4*x2**2 - 4)*x2**2 )
    if print_vals == (2,):
        print(f'Current x1: {x1:.5} Current x2: {x1:.5} Current Q: {Q:.5}')
    return Q
```

```python
print('======= Rosenbrock function =======')
bounds = ((-1, 3), (-1, 3)) 
res = differential_evolution(lambda x: rosenbrock(x, print_vals = False), bounds)
print(f"res: {res.x} \nQ: {res.fun}")
yopt = rosenbrock(res.x, print_vals = False)
print(f'yopt: {yopt}')
```

```python
x1 = np.linspace(-1, 3, 20)
x2 = np.linspace(-1, 3, 20)
X1, X2 = np.meshgrid(x1, x2)
Y1 = rosenbrock(np.array([X1, X2]), print_vals = False)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(X1, X2, Y1, cmap = 'jet', alpha = .5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.scatter(*res.x, yopt, label = 'Optimum')
ax.legend()
plt.show()
```

```python
print('\n======= Six-hump camel back function =======')
bounds = ((-1.9, 1.9), (-1.1, 1.1))
res = differential_evolution(lambda x: six_humps(x, print_vals = False), bounds)
yopt = six_humps(res.x, print_vals = False)
x1 = np.linspace(-1.9*1.2, 1.9*1.2, 20)
x2 = np.linspace(-1.1*1.2, 1.1*1.2, 20)
X1, X2 = np.meshgrid(x1, x2)
Y2 = six_humps(np.array([X1, X2]), print_vals = False)
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(projection = '3d')
ax1.plot_surface(X1, X2, Y2, cmap = 'jet', alpha = .5)
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.scatter(*res.x, yopt, label = 'Optimum')
ax1.legend()
```
