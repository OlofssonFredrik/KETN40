```python
from scipy.optimize import linprog
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = np.array([-1, 1])
A = np.array([[-2, 1],
                [1, -3],
                [1, 1],
                [-1, 0],
                [0, -1]
                ])
b = np.array([2, 2, 4, 0, 0])
res = linprog(c,
            A_ub = A,
            b_ub = b
            )
print(res)
```

```python
y_opt = c @ res.x
print(f"Optimal value: {y_opt}")
```

```python
x1_ar = np.linspace(-10, 10, 10)
x2_ar1 = 2*x1_ar + 2
x2_ar2 = (x1_ar - 2)/3
x2_ar3 = -x1_ar + 4
fig, ax = plt.subplots()
ax.plot(x1_ar, x2_ar1, label = 'Constraint 1')
ax.plot(x1_ar, x2_ar2, label = 'Constraint 2')
ax.plot(x1_ar, x2_ar3, label = 'Constraint 3')
ax.axvline(color = 'k', label = 'Constraint 4')
ax.axhline(color = 'k', label = 'Constraint 5')
ax.scatter(*res.x, label = 'Optimum', color = 'r')
ax.set_xlim(left = -1, right = 5)
ax.set_ylim(bottom = -1, top = 5)
ax.set_xlabel(r'x$_1$')
ax.set_ylabel(r'x$_2$')
ax.legend()
```
