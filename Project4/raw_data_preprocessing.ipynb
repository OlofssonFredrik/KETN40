```python
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

```python
#Open images in path water/img

path = 'water/'
files = os.listdir(path)

print(f"Number of images in {path} is {len(files)}")

#Check the unique format of the images
formats = set()
for file in files:
    formats.add(file.split('.')[-1])

print(f"Unique formats of the images are {formats}")
```

```python
path = Path('water')
desired_formats = {'jpg', 'jpeg', 'png'}
formats = set()

for file_name in os.listdir(path):
    file_path = path / file_name
    
    if file_path.is_file():
        file_format = file_path.suffix[1:] 
        formats.add(file_format)
        
        if file_format.lower() not in desired_formats:
            try:
                os.remove(file_path)
                print(f"Deleted: {file_path}")
            except Exception as e:
                print(f"Error deleting {file_path}: {str(e)}")

# Output the findings
print(f"Number of images in {path} after deletion is {len(list(os.listdir(path)))}")
print(f"Remaining unique formats of the images are {formats}")
```

```python
files = [path / file_name for file_name in os.listdir('water')]

if files:
    img = plt.imread(files[0])
    plt.imshow(img)
    plt.axis(False)
    plt.show()
else:
    print("No files found in the specified path.")
```
